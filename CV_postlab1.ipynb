```python
#Q4 Take any color. Split the color channels. Observe the output of each of the individual channels.
```

```python
imgChan= cv2.imread("/content/COLORBALLS.jpg")

cv2_imshow(imgChan)
```

```python
blue_image = np.zeros_like(imgChan)
green_image = np.zeros_like(imgChan)
red_image = np.zeros_like(imgChan)
blue_image[:, :, 0] = imgChan[:, :, 0]  # Blue channel
green_image[:, :, 1] = imgChan[:, :, 1]  # Green channel
red_image[:, :, 2] = imgChan[:, :, 2]  # Red channel
```

```python
cv2.cvtColor(blue_image, cv2.COLOR_BGR2RGB)
```

```python
cv2.cvtColor(green_image, cv2.COLOR_BGR2RGB)
```

```python
cv2.cvtColor(red_image, cv2.COLOR_BGR2RGB)
```

```python
#Q5 Invert all the output images obtained in previous task
```

```python
inverted_blue = 255 - blue_image
inverted_green = 255 - green_image
inverted_red = 255 - red_image

cv2.cvtColor(inverted_blue, cv2.COLOR_BGR2RGB)

```

```python
cv2.cvtColor(inverted_green, cv2.COLOR_BGR2RGB)
```
