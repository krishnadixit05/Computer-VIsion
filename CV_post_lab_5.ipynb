```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image and template
img = cv2.imread("/content/Img1.jpg", 0)  # Load as grayscale
template = cv2.imread("/content/lamp.png", 0)  # Load as grayscale

# Resize if necessary
img = cv2.resize(img, (0, 0), fx=0.8, fy=0.8)
template = cv2.resize(template, (0, 0), fx=0.8, fy=0.8)

# Perform template matching
method = cv2.TM_CCOEFF_NORMED
result = cv2.matchTemplate(img, template, method)

# Get the best match location
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
location = max_loc

# Define threshold for better accuracy
threshold = 0.8

# Check if the match is above the threshold
if max_val >= threshold:
    # Get template dimensions
    h, w = template.shape

    # Calculate bounding box coordinates
    bottom_right = (location[0] + w, location[1] + h)

    # Draw the bounding box
    cv2.rectangle(img, location, bottom_right, (255, 255, 255), 5)

    # Display the result
    cv2_imshow(img)
else:
    print("Template not found with sufficient confidence.")
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image and template
img = cv2.imread("/content/Img1.jpg", 0)  # Load as grayscale
template = cv2.imread("/content/lamp.png", 0)  # Load as grayscale

# Resize if necessary
img = cv2.resize(img, (0, 0), fx=0.8, fy=0.8)
template = cv2.resize(template, (0, 0), fx=0.8, fy=0.8)

# Perform template matching
method = cv2.TM_CCOEFF_NORMED
result = cv2.matchTemplate(img, template, method)

# Get the best match location
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
location = max_loc

# Define threshold for better accuracy
threshold = 0.8

# Check if the match is above the threshold
if max_val >= threshold:
    # Get template dimensions
    h, w = template.shape

    # Calculate bounding box coordinates
    bottom_right = (location[0] + w, location[1] + h)

    # Draw the bounding box and fill it with color
    cv2.rectangle(img, location, bottom_right, (255, 255, 255), -1)  # -1 fills the rectangle

    # Display the result
    cv2_imshow(img)
```
