```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros(shape=(500,500), dtype = np.int32)
plt.imshow(img, cmap='gray')
```

```python
cv2.putText(img,"GOOgle",(100,400), cv2.FONT_HERSHEY_PLAIN, fontScale= 5, color=(255,255,255), thickness=10)
plt.imshow(img, cmap='gray')
```

```python
img1 = np.zeros(shape=(500,500), dtype = np.int32)
plt.imshow(img1, cmap='gray')
```

```python
cv2.putText(img1, "Future", (100,400), cv2.FONT_HERSHEY_PLAIN, fontScale=5, color=(255,255,255), thickness=20)
cv2.putText(img1, "Future", (100,400), cv2.FONT_HERSHEY_PLAIN, fontScale=5, color=(0,0,0), thickness=10)
plt.imshow(img1, cmap='gray')
```
